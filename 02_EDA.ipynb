```python
columns = ames_clean.columns.drop('Unnamed: 0')
 
# Plot distributions for the selected columns
for column in columns:
    plt.figure(figsize=(8, 6))
    plt.hist(ames_clean[column], bins=20, color='skyblue', edgecolor='black')  # Adjust bins as needed
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
 
```

```python
for column in columns:
    plt.figure(figsize=(8, 6))
    plt.scatter(ames_clean[column], ames_clean['saleprice'], color='skyblue', edgecolor='black')
    plt.title(f'{column} vs SalePrice')
    plt.xlabel(column)
    plt.ylabel('SalePrice')
    plt.grid(True)
    plt.show()
```
